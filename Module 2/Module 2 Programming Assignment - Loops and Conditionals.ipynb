```python
secret = 2
guess = 8
if guess < secret:
    print("too low")
elif guess > secret:
    print("too high")
else:
    print("just right")
```

```python
small = False
green = False

if small and green:
    print("pea")
elif small and not green:
    print("cherry")
elif not small and green:
    print("watermelon")
elif not small and not green:
    print("pumpkin")
```

```python
guess_me = 7
number = 4

while number <= guess_me:
    if number < guess_me:
        print("too low")
    elif number == guess_me:
        print("found it!")
        break
    elif number > guess_me:
        print("oops")
        break

    number += 1
```

```python
guess_me = 5

for number in range(10):
    if number < guess_me:
        print("too low")
    elif number == guess_me:
        print("found it!")
        break
    elif number > guess_me:
        print("oops")
        break
```
